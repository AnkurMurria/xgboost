```python
import optuna
import xgboost as xgb
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/Admission_Prediction.csv')
```

```python
df
```

```python
df['GRE Score'] = df['GRE Score'].fillna(df['GRE Score'].mean())
df['TOEFL Score'] = df['TOEFL Score'].fillna(df['TOEFL Score'].mean())
df['University Rating'] = df['University Rating'].fillna(df['University Rating'].mean())
```

```python
x= df.drop(columns=['Serial No.','Chance of Admit'])
```

```python
y = df['Chance of Admit']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,random_state=42, test_size=0.3)
```

```python
model = xgb.XGBRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test, preds))
```

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
```

```python
rfc = RandomForestClassifier()
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, y_train)
```

```python
r_preds = rfr.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, r_preds))
```

```python
params = {
    'booster' : ['gbtree', 'gblinear'],
    'eta' : [0.1,0.2,0.3,0.5,1.0],
    'gamma': [0,0.1,0.2,1,2],
    'sample_type' : ['uniform', 'weighted']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
xgr = xgb.XGBRegressor()
```

```python
grid = GridSearchCV(xgr, param_grid=params, verbose=3, cv=5)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
x_preds = grid.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, x_preds))
```
